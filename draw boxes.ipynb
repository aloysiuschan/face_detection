```python
import cv2
import pandas as pd
from PIL import Image
import numpy as np
```

```python
full_labels = pd.read_csv('data/smile_labels.csv')
```

```python
full_labels.head()
```

```python
def draw_boxes(image_name):
    selected_value = full_labels[full_labels.filename == image_name]
    img = Image.open('images/{}'.format(image_name))
    img = np.array(img)
    for index, row in selected_value.iterrows():
        img = cv2.rectangle(img, (row['xmin'], row['ymin']), (row['xmax'], row['ymax']), (0, 255, 0), 3)
    return img
```

```python
Image.fromarray(draw_boxes(full_labels['filename'][0]))
```

```python
Image.fromarray(draw_boxes(full_labels['filename'][1]))
```
